```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn import linear_model  as lm
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.loc[(df['experience'].isnull()) & (df['interview_score(out of 10)'] == 9), 'experience'] = 'one'

print(df)
```

```python
# Update the 'experience' column where 'experience' is 'non' and 'interview_score' is '6'
df.loc[(df['experience'].isnull()) & (df['interview_score(out of 10)'] == 6), 'experience'] = 'zero'

print(df)
```

```python
df.rename(columns={'interview_score(out of 10)': 'interview_score'}, inplace=True)
df.rename(columns={'test_score(out of 10)': 'test_score'}, inplace=True)
df.rename(columns={'salary($)': 'salary'}, inplace=True)
df
```

```python
import math
test_median=df['test_score'].median()
```

```python
df['test_score'].fillna(test_median, inplace=True)
```

```python
from word2number import w2n
from word2number import w2n

# Convert each value in the 'experience' column to a numeric value
numeric_values = df['experience'].apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)

# Replace the 'experience' column with the converted numeric values
df['experience'] = numeric_values

# Example usage
print(df['experience'])
```

```python
reg=lm.LinearRegression()
```

```python
reg.fit(df[['experience','test_score','interview_score']],df['salary'])
```
